```python
# np.arange(start, end, step)

x = np.arange(0, 3*np.pi, 0.1)
y = np.sin(x)
plt.plot(x,y)
```

```python
ySin = np.sin(x)
yCos = np.cos(x)

plt.plot(x, ySin)
plt.plot(x, yCos)

plt.xlabel('X axis label')
plt.ylabel('Y axis label')
plt.title('Sine & Cosine Functions')
#ply.show()
```

```python
plt.subplot(2,1,1) # 2 rows, 1 column, position 1

plt.plot(x, ySin)
plt.title('Sine')

plt.subplot(2,1,2) # 2 rows, 1 column, position 2
plt.plot(x, yCos)
plt.title('Cosine')
```

```python
path = 'datasets/Cartwheeldata.csv'
df = pd.read_csv(path)
df.head(2)
```

```python
sns.lmplot(x = 'Wingspan', 
           y = 'Height',
           data = df,
           #fit_reg = False
           hue = 'Gender'
          )
plt.title('Scatter plot of Wingspans vs Height')
```

```python
sns.swarmplot(x = 'Gender',
              y = 'Wingspan',
              data = df
             )
plt.title('Gender vs Wingspan, the Swarmplot')
```

```python
sns.boxplot(data = df.loc[:,['Age', 'Height', 'Wingspan', 'CWDistance']])
                          
```

```python
sns.boxplot(data = df.loc[df['Gender'] == 'F',
                             ['Age', 'Height', 'Wingspan', 'CWDistance']])
```

```python
sns.distplot(df.CWDistance,
            #kde = False
            )
```

```python
sns.countplot(x = 'Gender',
              data = df)
plt.xticks(rotation = 45)
plt.yticks(rotation = 45)
```
